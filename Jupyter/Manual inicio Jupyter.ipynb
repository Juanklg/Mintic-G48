# Como ejecutar Notebook de Jupyter

### 1 - verificar los entornos(env) que hemos creado

Ejecutamos en terminal el siguiente comando

##### comando 1 ->  conda env list

Muestra los entornos(env) creados y con un asterisco(*) nos indica cual es el entornos(env) activo

Si no hemos creado el env donde vamos a trabajar entonces damos el siguiente paso

### 2 - Crear un entornos(env)

##### comando 2 -> conda create -n jupyter

Nos va a preguntar si queremos proceder a crear el env y le decimos q si (y). 
Despues verificamos de nuevo si se creo el env con el comando 1

##### comando 1 ->  conda env list

Si ya esta creado, procedemos a activar el nuevo env que creamos en el paso 3

### 3 - Activar un entorno(env)

Se activa con el siguiente comando, en este caso jupyter es el nombre del env
##### comando 3 -> conda activate jupyter

puede que me diga que no esta activado el shell en conda entonces le damos el comando 4
##### comando 4 -> conda init bash

verificamos que este activo con el comando 1 y con el asterisco indicar que estamos en el env deseado. Verificamos si esta instalado el notebook con el comando 5
##### comando 5 -> conda list

si no hemos instalado notebook vamos al paso 4

### 4 - Instalar Notebook de JupyterLab en env Jupyter

instalamos con el comando 6, en este caso se instala desde la libreria de conda-forge
##### comando 6 -> conda install -c conda-forge notebook
Me preguntara si deseo instalar toda la lista de paquetes a lo cual damos si(y)

Verificamos que si se instalo el notebook con el comando 7
##### comando 7 -> conda list


### 5 - Iniciar Jupyter Notebook
Primero nos unbicamos en la carpeta donde queremos ejecutar jupiter notebook (1)

Con el comando "ls" podemos observar las carpetas que hay dentro de la ruta que seleccionamos

cuando nos aseguremos que estamos ubicados sobre la ruta que deseamos trabajar ejecutamos el comando 8

##### comando 8 -> jupyter notebook


Se iniciara el jupyter notebook en nuestro navegador o podemos dar "ctrl + click" en la direccion que nos muestra la terminal
